```julia
using Revise, Statistics,Distributions, LaTeXStrings, Plots
plotlyjs(); default(background_color=:transparent, foreground_color="#d8dee9")
```

```julia

# Task 1 illustration
xs = range(-15,30,200)
θ = 15.0; σ = 5
f1 = Normal(0, σ)
f2 = Normal(θ, σ)
plot(xs, pdf.(f1, xs), label="Noise", ticks=nothing)
plot!(xs, pdf.(f2, xs), c=:orange,label="Signal+Noise", ticks=nothing)
plot!([θ,θ],[pdf(f2,θ),0], c=:orange, legend=false, line=(1,:dash), framestyle=:zerolines)
ylims!(ylims().*[1.1,1.0]...)
annotate!(θ, -0.0026, "θ")
annotate!(-0.5, -0.0026, "0")
annotate!(3.5, pdf(f1,0), "Noise", current().series_list[1].plotattributes[:linecolor])
annotate!(θ+6.5, pdf(f2,θ), "Signal+Noise", current().series_list[2].plotattributes[:linecolor])
savefig("PDFs.pdf")
current()
```
